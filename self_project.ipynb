## Filling the NA values[IMP]

```r
# using any ordinary methods like df|>fil(...) or using zoo package to na.locb, or na.nocf ie backfill and frontfill will fill the NA values with a single value resulting in a strong biased filling

# As a measure we are willing to fill the values with values withing a certain range thus making a more distributed filling
```

```r
# The runif method generates random values provided it gets the number to generate random values with range in min and max value

# runif(number of values to generate, min_value, max_value)
```

```r
min_value <- 138
max_value <- 177

random_values<- runif(sum(is.na(daily_mean_AQI$mean_AQI)), min=min_value, max=max_value)
```

```r
sum(is.na(daily_mean_AQI$mean_AQI))
```

```r
random_values
```

```r
daily_mean_AQI$mean_AQI[is.na(daily_mean_AQI$mean_AQI)]<-random_values[1:(sum(is.na(daily_mean_AQI$mean_AQI)))]
```

```r
daily_mean_AQI|>autoplot()
```

```r
 # Hence we have generated some random values in the range and filled the NA values with that and thus makes the data more randomized and natural
```

```r
ACF(daily_mean_AQI,mean_AQI)|>autoplot()
```
